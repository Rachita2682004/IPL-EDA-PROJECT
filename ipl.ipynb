Maximum wickets taken in each season along with bwoler name

```python
valid_dismissals = ["caught", "lbw", "bowled", "caught and bowled", "stumped", "hit wicket"]
bowler_data = matches_and_deliveries[matches_and_deliveries["dismissal_kind"].isin(valid_dismissals)]

wickets_per_bowler = bowler_data.groupby(["season", "bowler"]).size().reset_index(name="wickets")

most_wickets_per_season = wickets_per_bowler.loc[wickets_per_bowler.groupby("season")["wickets"].idxmax()].reset_index(drop=True)

most_wickets_per_season = most_wickets_per_season.sort_values("season").reset_index(drop=True)
```

```python
most_wickets_per_season
```

Shown : Which bowler has taken max wickets per season along with no. of wickets taken.

```python
plt.figure(figsize=(15,8))
max_wickets = sns.barplot(x = "season", y = "wickets", data = most_wickets_per_season , hue = "bowler")
for i in max_wickets.containers:
    plt.bar_label(i)
plt.ylabel("No. of wickets taken")
plt.xticks(rotation = 90)
plt.show()
```

Number of dot balls in every season

```python
valid_data = matches_and_deliveries[(matches_and_deliveries["batsman_runs"] == 0) & (matches_and_deliveries["extra_runs"] == 0) & (matches_and_deliveries["total_runs"]== 0)]

dot_bowl_per_season = valid_data.groupby(["season", "bowler"]).size().reset_index(name="dot_balls")

most_dot_bowls_per_season = dot_bowl_per_season.loc[dot_bowl_per_season.groupby("season")["dot_balls"].idxmax()].reset_index(drop=True)

most_dot_bowls_per_season = most_dot_bowls_per_season.sort_values("season").reset_index(drop=True)
```

```python
most_dot_bowls_per_season
```

Shown : Which bowler has given max dot balls per season along with no. of dot balls.

```python
plt.figure(figsize=(15,8))
max_dot_balls = sns.barplot(x = "season", y = "dot_balls", data = most_dot_bowls_per_season , hue = "bowler")
for i in max_dot_balls.containers:
    plt.bar_label(i)
plt.ylabel("No. of dot balls")
plt.xticks(rotation = 90)
plt.show()
```
